# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
        print("vehicle nesnesi türetildi")
    def intro(self):
        print(self.name,self.speed,self.mileage)
class Bus(Vehicle):
    def __init__(self, name, speed, mileage,capacity):
        super().__init__(name, speed, mileage)
        self.capacity=capacity
        print("bus nesnesi türetildi")
    def intro(self):
        print(self.name,self.speed,self.mileage,self.capacity)
    def seat(self):
        print("The seating capacity of a bus is {} passengers".format(self.capacity))
````

````python
bus1=Bus("Volvo",180,12,50)
bus1.intro()
bus1.seat()
````

````python
class Vehicle:
    pass

class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        self.capacity = capacity
        print(f"The seating capacity of a bus is {self.capacity} passengers.")
````

````python
class Vehicle:
    pass

class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        self.capacity = capacity
        print(f"This bus has a seating capacity of {self.capacity} passengers.")

class School_bus(Bus):
    pass

# Create an instance of School_bus and check if it is also an instance of Vehicle
my_school_bus = School_bus()
print(isinstance(my_school_bus, Vehicle))  # True

# Call seating_capacity method of Bus class using an instance of School_bus
my_school_bus.seating_capacity() # This bus has a seating capacity of 50 passengers.
````
