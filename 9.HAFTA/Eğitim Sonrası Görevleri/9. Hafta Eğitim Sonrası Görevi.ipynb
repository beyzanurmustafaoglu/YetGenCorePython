```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durumu = False
        self.ses = 0
        self.kanal_listesi = ["TRT", "CNN", "BBC", "FOX"]
        self.kanal = "TRT"
        
    def tv_ac(self):
        if self.tv_durumu:
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            self.tv_durumu = True
        
    def tv_kapat(self):
        if self.tv_durumu:
            print("Televizyon kapatılıyor...")
            self.tv_durumu = False
        else:
            print("Televizyon zaten kapalı.")
            
    def ses_ayarları(self):
        while True:
            cevap = input("Sesi azaltmak için '<', artırmak için '>', çıkmak için 'q' tuşuna basın: ")
            if cevap == "<":
                if self.ses != 0:
                    self.ses -= 1
                    print("Ses:", self.ses)
            elif cevap == ">":
                if self.ses != 31:
                    self.ses += 1
                    print("Ses:", self.ses)
            else:
                print("Ses güncellendi:", self.ses)
                break
                
    def kanal_ekle(self, kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print("Kanal eklendi.")
        
    def kanal_sayisi(self):
        print("Kanal sayısı:", len(self.kanal_listesi))
        
    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele]
        print("Şu anki kanal:", self.kanal)
        
    def televizyon_bilgileri(self):
        print("Tv durumu:", self.tv_durumu)
        print("Ses:", self.ses)
        print("Kanal listesi:", self.kanal_listesi)
        print("Şu anki kanal:", self.kanal)
        
kumanda = Kumanda()

while True:
    print("""\nTelevizyon Kumandası:
    1. Tv aç
    2. Tv kapat
    3. Ses ayarları
    4. Kanal ekle
    5. Kanal sayısını öğrenme
    6. Rastgele kanala geçme
    7. Televizyon bilgileri
    Çıkmak için q'ya basın.""")
    islem = input("İşlem seçiniz: ")
    
    if islem == "q":
        print("Kumanda kapatılıyor...")
        break
        
    elif islem == "1":
        kumanda.tv_ac()
        
    elif islem == "2":
        kumanda.tv_kapat()
        
    elif islem == "3":
        kumanda.ses_ayarları()
        
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak gir")
```
