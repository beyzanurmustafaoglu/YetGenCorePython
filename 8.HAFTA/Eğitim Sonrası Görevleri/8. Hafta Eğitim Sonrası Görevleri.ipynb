 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# Kullanıcı giriş sayısını al
giris_sayisi = int(input("Kaç giriş yapacaksınız? "))

# Verileri saklamak için boş sözlük oluştur
veriler = {}

# Her giriş için kullanıcı verilerini al
for i in range(giris_sayisi):
    # Kullanıcı verilerini al
    isim_soyisim = input("Adınız soyadınız: ")
    yas = int(input("Yaşınız: "))
    boy = float(input("Boyunuz (metre cinsinden): "))
    kilo = float(input("Kilonuz (kg cinsinden): "))
    
    # VKİ hesapla
    vki = kilo / (boy ** 2)
    
    # VKİ'yi sınıflara göre kaydet
    if vki < 18.5:
        sinif = "Zayif"
    elif 18.5 <= vki < 25:
        sinif = "Normal kilolu"
    elif 25 <= vki < 30:
        sinif = "Fazla kilolu"
    elif 30 <= vki < 40:
        sinif = "Obez"
    else:
        sinif = "İleri derecede obez (morbid obez)"
    
    # Verileri sözlüğe kaydet
    veriler[isim_soyisim] = {"Yas": yas, "Boy": boy, "Kilo": kilo, "VKI": vki, "Sinif": sinif}

# Verileri dosyaya kaydet
with open("kullanici_verileri.txt", "w") as dosya:
    for isim_soyisim, veri in veriler.items():
        dosya.write(f"{isim_soyisim}: {veri}\n")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
```

```python
import datetime

# İçecek menüsü
menu = {
    "Acai Smoothie": 8.99,
    "Kale Smoothie": 7.99,
    "Protein Shake": 9.99,
    "Green Juice": 6.99,
    "Coconut Water": 3.99
}

# İçecek fiyatlarını hesapla
def calculate_price(item, quantity):
    price = menu.get(item) * quantity
    return price

# Satın alma işlemi yap
def buy():
    # Kullanıcıya menüyü göster
    print("----- İçecek Menüsü -----")
    for item in menu:
        print(f"{item}: ${menu[item]}")
    
    # Kullanıcıdan içecek seçimini al
    item = input("Lütfen içecek seçiminizi yapın: ")
    
    # Kullanıcıdan içecek adedi seçimini al
    quantity = int(input("Lütfen adet girin: "))
    
    # Fiyat hesapla
    price = calculate_price(item, quantity)
    
    # Tarih bilgisi
    date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Satın alma bilgilerini yazdır
    print(f"{quantity} adet {item} seçtiniz. Toplam fiyat: ${price}")
    
    # Satın alma bilgilerini kaydet
    with open("log.txt", "a") as file:
        file.write(f"{date} - {quantity} adet {item} satın alındı. Fiyat: ${price}\n")

# Menüyü göster
def show_menu():
    print("1. Menüyü görüntüle")
    print("2. Satın al")
    print("3. Çıkış")

# Ana program
while True:
    show_menu()
    choice = input("Lütfen bir seçim yapın: ")
    
    if choice == "1":
        # Menüyü göster
        print("----- İçecek Menüsü -----")
        for item in menu:
            print(f"{item}: ${menu[item]}")
    elif choice == "2":
        # Satın alma işlemi yap
        buy()
    elif choice == "3":
        # Çıkış yap
        print("Teşekkür ederiz. İyi günler!")
        break
    else:
        # Hatalı seçim
        print("Geçersiz seçim. Lütfen tekrar deneyin.")
```
